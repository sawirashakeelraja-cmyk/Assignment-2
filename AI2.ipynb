```python
# Install required libraries (Run only once in Google Colab)
!pip -q install nltk

import nltk
import re

# Download required datasets
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
text = "I love Artificial Intelligence because it is making our lives better by playing an important role."

print("="*60)
print("STEP 1 : INPUT TEXT")
print("="*60)
print(text)
```

```python
text = text.lower()

text = re.sub(r'[^a-zA-Z\s]', '', text)

print("\n" + "="*60)
print("STEP 2 : TEXT PREPROCESSING")
print("="*60)
print(text)
```

```python
tokens = text.split()

print("\n" + "="*60)
print("STEP 3 : TOKENIZATION")
print("="*60)

print(tokens)
```

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = []

for word in tokens:

    if word not in stop_words:

        filtered_tokens.append(word)

print("\n" + "="*60)
print("STEP 4 : STOP WORD REMOVAL")
print("="*60)

print(filtered_tokens)
```

```python
stemmer = PorterStemmer()

stemmed_words = []

for word in filtered_tokens:

    stemmed_words.append(stemmer.stem(word))

print("\n" + "="*60)
print("STEP 5 : STEMMING")
print("="*60)

print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = []

for word in filtered_tokens:

    lemmatized_words.append(lemmatizer.lemmatize(word))

print("\n" + "="*60)
print("STEP 6 : LEMMATIZATION")
print("="*60)

print(lemmatized_words)
```

```python
print("\n" + "="*60)
print("FINAL OUTPUT OF NLP PREPROCESSING")
print("="*60)

print("Original Sentence : ", text)

print()

print("Tokens : ", tokens)

print()

print("Without Stopwords : ", filtered_tokens)

print()

print("Stemmed Words : ", stemmed_words)

print()

print("Lemmatized Words : ", lemmatized_words)
```

```python
# Install Sentence Transformers
!pip -q install sentence-transformers

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
print("="*60)
print("STEP 7 : LOADING AI MODEL")
print("="*60)

model = SentenceTransformer('all-MiniLM-L6-v2')

print("AI Model Loaded Successfully!")
```

```python
sentence1 = "Artificial Intelligence makes our lives easier."
sentence2 = "AI helps people perform tasks more efficiently."

print("\n" + "="*60)
print("STEP 8 : INPUT SENTENCES")
print("="*60)

print("Sentence 1:", sentence1)
print("Sentence 2:", sentence2)

embedding1 = model.encode(sentence1)
embedding2 = model.encode(sentence2)

print("\nEmbedding 1 Shape:", embedding1.shape)
print("Embedding 2 Shape:", embedding2.shape)

print("\nFirst 10 Values of Embedding 1:")
print(embedding1[:10])

print("\nFirst 10 Values of Embedding 2:")
print(embedding2[:10])
```

```python
print("\n" + "="*60)
print("STEP 9 : COSINE SIMILARITY")
print("="*60)

similarity = cosine_similarity(
    [embedding1],
    [embedding2]
)

print("Similarity Score:")
print(similarity)
```

```python
print("\n" + "="*60)
print("STEP 10 : AI MODEL INFERENCE")
print("="*60)

score = similarity[0][0]

if score > 0.80:
    prediction = "Highly Similar"

elif score > 0.60:
    prediction = "Moderately Similar"

else:
    prediction = "Different Meaning"

print("Prediction:")
print(prediction)
```

```python
print("\n" + "="*60)
print("FINAL NLP PIPELINE")
print("="*60)

print("Original Sentence 1:")
print(sentence1)

print("\nOriginal Sentence 2:")
print(sentence2)

print("\nEmbedding Size:")
print(len(embedding1))

print("\nSimilarity Score:")
print(score)

print("\nFinal Prediction:")
print(prediction)
```
